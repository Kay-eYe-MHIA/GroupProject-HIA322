```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("D:/IMU/Health Data Collection HIA322/Project & Assignments/Group Project/cleaned_data.csv")

# Select target variable and predictor
target = data['cases_new']
predictor = data['checkins']

# Add a constant to the predictor for regression analysis
X = sm.add_constant(predictor)
model = sm.OLS(target, X).fit()

# Print summary of regression
print(model.summary())

# Plot the trend between check-ins and new cases
plt.figure(figsize=(10, 6))
plt.scatter(predictor, target, alpha=0.5)
plt.plot(predictor, model.predict(X), color='red', label='Regression Line')
plt.title('Trend Analysis: Check-ins vs New Cases')
plt.xlabel('Check-ins')
plt.ylabel('New Cases')
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm

# Define predictors and two dependent variables
X = sm.add_constant(data['checkins'])
y1 = data['cases_new']
y2 = data['cases_active']

# Regression for cases_new
model_cases_new = sm.OLS(y1, X).fit()

# Regression for cases_active
model_cases_active = sm.OLS(y2, X).fit()

# Print summaries
print("Regression Results for cases_new")
print(model_cases_new.summary())

print("\nRegression Results for cases_active")
print(model_cases_active.summary())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("D:/IMU/Health Data Collection HIA322/Project & Assignments/Group Project/cleaned_data.csv")  # Replace with your file path

# Select relevant columns (cluster columns, check-ins, and cases_new)
columns_of_interest = ['checkins', 'cases_new', 'cluster_import', 'cluster_religious', 
                       'cluster_community', 'cluster_highRisk', 'cluster_education', 
                       'cluster_detentionCentre', 'cluster_workplace']
subset_data = data[columns_of_interest]

# Compute correlation matrix
correlation_matrix = subset_data.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt

# Step 3: Export the heatmap
plt.title('Correlation Heatmap')
plt.savefig('correlation_heatmap.png')  # Save as PNG file
plt.show()
```

```python
import statsmodels.api as sm

# Define predictors (check-ins + cluster columns) and dependent variable (cases_new)
X = data[['checkins', 'cluster_import', 'cluster_religious', 'cluster_community', 
          'cluster_highRisk', 'cluster_education', 'cluster_detentionCentre', 
          'cluster_workplace']]
X = sm.add_constant(X)  # Add constant for intercept
y = data['cases_new']

# Fit OLS model
model = sm.OLS(y, X).fit()

# Print summary
print(model.summary())
```
